### Чистые функции

1. **Неизменяемость (Immutability)**
Чистые функции не изменяют входные данные. Вместо изменения состояния объектов они возвращают новые объекты. Это свойство делает код более безопасным и легким для отладки, поскольку состояние системы остается неизменным между вызовами функций.

2. **Без побочных эффектов (No Side Effects)**
Чистые функции не влияют на внешний мир за пределами своих параметров и возвращаемого значения. Они не взаимодействуют с глобальным состоянием, не изменяют внешние переменные и не вызывают сторонних эффектов, таких как ввод-вывод, сетевые запросы или модификация базы данных.

3. **Предсказуемость (Determinism)**
Из-за отсутствия побочных эффектов и неизменяемости, результаты вызова чистой функции всегда одинаковы при одних и тех же входных данных. Это свойство делает тестирование и отладку кода гораздо проще, поскольку можно быть уверенным в том, что функция будет работать одинаково в разных средах и с разными данными.

4. **Пересчитываемость (Referential Transparency)**
Если функция является чистой, её результат можно заменить на её аргументы без изменения поведения программы. Это означает, что любое место в коде, где используется результат функции, может быть заменено непосредственным использованием этого результата, что упрощает оптимизацию и анализ кода.

    Чистая? 

````python
def add(a, b):
    return a + b

print(add(4, 5))
print(add(4, 5))
print(add(4, 5))
````

- %u: беззнаковое целое число (unsigned integer)
- %U: беззнаковое целое число в верхнем регистре (unsigned integer)
- %c: символ (character)
- %r: репрезентация строки (string representation)
- %q: квантованная строка (quotation)

## Кодирование

### 1. Кодирование текста
Кодирование текста - это процесс представления символов в виде числовых значений, которые могут быть обработаны компьютером. Существует несколько стандартов кодирования, каждый из которых имеет свои особенности и область применения.

### 2. ASCII (American Standard Code for Information Interchange)
- **История и назначение**: ASCII был разработан в 1960-х годах как стандарт для представления текстовых символов в компьютерах и других устройствах.
- **Диапазон символов**: ASCII использует 7 бит для кодирования, что позволяет закодировать 128 символов (от 0 до 127). Это включает латинские буквы (A-Z, a-z), цифры (0-9), знаки препинания и специальные управляющие символы (например, переход на новую строку).
- **Ограничения**: ASCII подходит только для английского языка и не поддерживает символы других языков или специальные знаки.


### 3. Unicode
- **История и назначение**: Unicode был разработан для решения проблемы ограниченности ASCII, обеспечивая универсальный стандарт кодирования для всех языков мира.
- **Диапазон символов**: Unicode поддерживает более 143,000 символов, охватывая множество языков, включая азиатские, арабские и другие письменности, а также специальные символы и эмодзи.
- **Кодовые пространства**: Unicode использует несколько форм представления, включая UTF-8, UTF-16 и UTF-32, для различных потребностей и ограничений.

### 4. UTF-8 (Unicode Transformation Format - 8-bit) - метод кодирования 
- **Описание**: UTF-8 является одним из самых популярных методов кодирования Unicode символов. Он оптимизирован для использования в вебе и других приложениях, где важна совместимость и экономия памяти.
- **Преимущества**:
  - **Совместимость с ASCII**: Первые 128 символов UTF-8 совпадают с ASCII, что обеспечивает обратную совместимость.
  - **Эффективность**: UTF-8 использует переменную длину кодирования: символы ASCII кодируются 1 байтом, а другие символы могут занимать от 2 до 4 байтов.
  - **Гибкость**: UTF-8 может кодировать любой символ Unicode, обеспечивая широкую поддержку различных языков и символов.
- **Применение**: Широко используется в веб-разработке, операционных системах и других приложениях, где важна поддержка мультиязычного контента.


### Другие методы кодирования Unicode

1. **UTF-16 (Unicode Transformation Format - 16-bit)**
- **Описание**: UTF-16 использует 16-битные (2-байтовые) единицы для кодирования символов.
- **Структура**: Символы из базовой многоязычной плоскости (BMP, U+0000 до U+FFFF) кодируются одним 16-битным словом. Символы вне BMP (U+10000 до U+10FFFF) кодируются двумя 16-битными словами (суррогатная пара).
- **Преимущества**: Эффективен для текстов, содержащих много не-ASCII символов.
- **Недостатки**: Увеличивает объем данных для текстов, состоящих в основном из символов ASCII.

2. **UTF-32 (Unicode Transformation Format - 32-bit)**
- **Описание**: UTF-32 использует фиксированную длину 32-битных (4-байтовых) единиц для кодирования символов.
- **Структура**: Каждый символ Unicode представляется ровно одним 32-битным словом.
- **Преимущества**: Простота обработки, так как каждый символ занимает одинаковое количество байтов.
- **Недостатки**: Неэффективен по памяти, так как каждый символ занимает 4 байта, независимо от его кодовой точки.

### Методы кодирования для других целей

1. **ISO/IEC 8859 (Latin-1)**
- **Описание**: Набор стандартов, разработанных для кодирования символов латинских алфавитов.
- **Структура**: Каждый стандарт из серии (например, ISO/IEC 8859-1) использует 8 бит для кодирования, что позволяет представлять 256 символов.
- **Применение**: Используется для европейских языков до широкого внедрения Unicode.


2. **CP1252 (Windows-1252)**
- **Описание**: Однобайтовая кодировка, используемая в операционных системах Windows для западноевропейских языков.
- **Структура**: Похожа на ISO/IEC 8859-1, но с добавлением дополнительных символов в диапазоне 128-159.
- **Применение**: Широко использовалась в документах и веб-страницах до распространения UTF-8.


3. **EBCDIC (Extended Binary Coded Decimal Interchange Code)**
- **Описание**: Однобайтовая кодировка, разработанная IBM для использования на мэйнфреймах.
- **Структура**: Использует 8 бит для кодирования символов, но с другим набором значений, чем ASCII.
- **Применение**: Применяется в старых системах IBM и мэйнфреймах.


4. **Shift JIS**
- **Описание**: Кодировка для японского языка, разработанная для совместимости с ASCII и расширением для японских символов.
- **Структура**: Использует переменную длину (1 или 2 байта) для кодирования символов.
- **Применение**: Широко используется в Японии для текстов на японском языке.


````python
print(ord("r")) # ASCII 
print(chr(ord("r"))) # обратное преобразование 

# "0x" указывает на 16ричное представление, 
# а "72" — это шестнадцатеричное значение числа 114.
print(hex(ord("r"))) 

print(b'\x72'.decode()) 
````

### Что такое байты?

**Байты** - это базовая единица цифровой информации, которая состоит из 8 битов. Бит может быть либо 0, либо 1, поэтому байт может представлять 256 различных значений (от 0 до 255). Байты используются для представления данных в компьютерах, включая текст, изображения, аудио и видео.

### Как Python читает байты?

В Python байты представлены типом данных `bytes`. Вот несколько способов, как Python работает с байтами:

1. **Чтение байтов из файла**:
   - Открытие файла в бинарном режиме с использованием `open()` с режимом `'rb'`.
   - Чтение байтов с помощью метода `read()`.

```python
with open('example.bin', 'rb') as file:
    byte_data = file.read()
    print(byte_data)
```

2. **Создание байтов вручную**:
   - Использование литералов байтов с префиксом `b`.
   - Преобразование строки в байты с использованием метода `encode()`.

```python
# Литералы байтов
byte_data = b'Hello, World!'
print(byte_data)

# Преобразование строки в байты
string_data = 'Hello, World!'
byte_data = string_data.encode('utf-8')
print(byte_data)
```

### Какие данные хранятся в байтах?

В байтах могут храниться различные типы данных, включая:

1. **Текстовые данные**:
   - Каждая буква или символ может быть закодирован в байты с помощью определенной кодировки (например, ASCII или UTF-8).

```python
# Кодирование строки в байты
string_data = 'Hello, World!'
byte_data = string_data.encode('utf-8')
print(byte_data)  # b'Hello, World!'
```

2. **Изображения**:
   - Графические файлы, такие как JPEG, PNG и GIF, хранят данные в виде байтов.

```python
with open('example.jpg', 'rb') as image_file:
    image_data = image_file.read()
    print(image_data[:10])  # Печать первых 10 байтов
```

3. **Аудио и видео**:
   - Медиафайлы, такие как MP3, WAV, MP4 и AVI, также хранят данные в виде байтов.

```python
with open('example.mp3', 'rb') as audio_file:
    audio_data = audio_file.read()
    print(audio_data[:10])  # Печать первых 10 байтов
```

4. **Двоичные данные**:
   - Любые другие данные, такие как двоичные файлы, исполняемые файлы (.exe), архивы (.zip), хранятся в виде байтов.

### Примеры работы с байтами в Python

1. **Преобразование байтов в строку**:
   - Использование метода `decode()`.

```python
byte_data = b'Hello, World!'
string_data = byte_data.decode('utf-8')
```

## Работа с файлами

### Открытие файла

Функция `open()` используется для открытия файла и возвращает объект файла. Синтаксис:
```python
file_object = open(filename, mode)
```

### Режимы открытия файла

- **'r'**: Чтение (по умолчанию). Файл должен существовать. **(+t по умолчанию)**
- **'w'**: Запись. Создает новый файл или обрезает существующий. **(+t по умолчанию)**
- **'a'**: Добавление. Создает новый файл или добавляет к существующему. **(+t по умолчанию)**
- **'b'**: Бинарный режим. Может комбинироваться с другими режимами ('rb', 'wb', 'ab').
- **'t'**: Текстовый режим (по умолчанию). Может комбинироваться с другими режимами ('rt', 'wt', 'at'). 
- **'+'**: Чтение и запись. Может комбинироваться с другими режимами ('r+', 'w+', 'a+').

Примеры:
```python
# Чтение текстового файла
with open('example.txt', 'r') as file:
    content = file.read()

# Запись в текстовый файл
with open('example.txt', 'w') as file:
    file.write('Hello, World!')

# Добавление в текстовый файл
with open('example.txt', 'a') as file:
    file.write('\nAppended text')

# Чтение бинарного файла
with open('example.jpg', 'rb') as file:
    binary_content = file.read()
```

### Методы чтения

- **`read(size)`**: Читает `size` байт или символов (если размер не указан, читает весь файл).
- **`readline()`**: Читает одну строку.
- **`readlines()`**: Читает все строки и возвращает список.

### Методы записи

- **`write(string)`**: Записывает строку в файл.
- **`writelines(list_of_strings)`**: Записывает список строк в файл.

### Курсор в файле

Курсор указывает на текущую позицию чтения или записи в файле. Его положение можно изменять и получать с помощью методов:

- **`tell()`**: Возвращает текущую позицию курсора.
- **`seek(offset, whence)`**: Перемещает курсор на `offset` байт от позиции `whence`.

Параметр `whence` может принимать следующие значения:
- **0**: От начала файла (по умолчанию).
- **1**: От текущей позиции.
- **2**: От конца файла.

Примеры:
```python
with open('example.txt', 'r') as file:
    print(file.tell())  # Позиция курсора (0)
    content = file.read(5)
    print(file.tell())  # Позиция курсора (5)
    
    file.seek(0)        # Перемещение курсора в начало
    print(file.tell())  # Позиция курсора (0)
    
    file.seek(5, 0)     # Перемещение курсора на 5 байт от начала
    print(file.tell())  # Позиция курсора (5)
    
    file.seek(-5, 2)    # Перемещение курсора на 5 байт от конца файла
    print(file.tell())  # Позиция курсора (конец файла - 5 байт)
```



**Важно всегда закрывать файл после завершения работы с ним, чтобы освободить ресурсы. Лучшей практикой является использование контекстного менеджера with, который автоматически закрывает файл даже в случае возникновения исключений.**

with - это контекстный менеджер 

    1 пример

````python
# Запись в файл
with open('output.txt', 'w') as file:
    file.write('Hello, world!')

# Чтение из файла
with open('output.txt', 'r') as file:
    content = file.read()
    print(content)
````

````python
my_list = ['apple', 'banana', 'cherry']

with open('fruits.txt', 'w') as file: # w - write a - append 
    for item in my_list:
        file.write(f"{item}\n")

with open('fruits.txt', 'r') as file:
    content = file.read()
    print(content)
````

### pickle (соленый огурец по полу валяется никто его не ест)
Pickle позволяет сериализовать практически любой объект Python, включая пользовательские классы. Это делает его очень мощным инструментом для сохранения сложных структур данных.

````python
import pickle

data = {
    'name': 'John Doe',
    'age': 30,
    'is_student': False
}

with open('data.pkl', 'wb') as file:
    pickle.dump(data, file)

with open('data.pkl', 'rb') as file:
    loaded_data = pickle.load(file)

print(loaded_data)
````

### json
JSON (JavaScript Object Notation) представляет собой открытый стандарт для обмена данными, который легко читается людьми и легко парсируется машинами. JSON поддерживает простые типы данных, такие как строки, числа, булевы значения, массивы и объекты. 

````python
import json

data = {
    'name': 'Jane Doe',
    'age': 28,
    'is_student': True
}

with open('data.json', 'w') as file:
    json.dump(data, file)

with open('data.json', 'r') as file:
    loaded_data = json.load(file)

print(loaded_data)
````

**Разница между Pickle и JSON**
- Типы данных: Pickle может сериализовать почти любой объект Python, включая пользовательские классы, в то время как JSON ограничивается простыми типами данных.
- Кросс-платформенность: JSON является кросс-платформенным форматом, который может быть прочитан большинством языков программирования, в то время как данные, сериализованные с помощью pickle, могут быть прочитаны только на Python.
- Безопасность: Из-за своей способности сериализовать произвольные объекты, pickle считается менее безопасным, чем JSON, и его следует использовать с осторожностью при работе с данными, полученными из ненадежных источников.

### Работа с файловой системой

1. **os.getcwd()**
- Возвращает текущий рабочий каталог.
   ```python
   import os
   print(os.getcwd())
   ```

2. **os.chdir(path)**
- Изменяет текущий рабочий каталог на указанный путь.
   ```python
   os.chdir('/path/to/directory')
   ```

3. **os.listdir(path='.')**
- Возвращает список файлов и директорий в указанном каталоге.
   ```python
   print(os.listdir('/path/to/directory'))
   ```

4. **os.mkdir(path, mode=0o777)**
- Создает новый каталог по указанному пути.
   ```python
   os.mkdir('/path/to/new_directory')
   ```

5. **os.makedirs(name, mode=0o777, exist_ok=False)**
- Создает промежуточные каталоги. Если `exist_ok=True`, ошибки при существовании каталогов не будет.
   ```python
   os.makedirs('/path/to/new_directory/nested_directory')
   ```

- 0o777 — это окталовая записи прав доступа. В этой системе:
- Первая цифра (7) относится к правам владельца каталога.
- Вторая цифра (7) относится к правам группы.
- Третья цифра (7) относится к правам всех остальных пользователей.

6. **os.remove(path)**
- Удаляет файл по указанному пути.
   ```python
   os.remove('/path/to/file')
   ```

7. **os.rmdir(path)**
- Удаляет пустой каталог по указанному пути.
   ```python
   os.rmdir('/path/to/directory')
   ```

8. **os.removedirs(name)**
- Удаляет каталоги рекурсивно.
   ```python
   os.removedirs('/path/to/directory/nested_directory')
   ```

9. **os.rename(src, dst)**
- Переименовывает файл или каталог.
   ```python
   os.rename('old_name', 'new_name')
   ```

10. **os.replace(src, dst)**
 - Переименовывает файл или каталог, заменяя целевой, если он существует.
    ```python
    os.replace('old_name', 'new_name')
    ```

11. **os.path.exists(path)**
 - Проверяет существование пути.
    ```python
    print(os.path.exists('/path/to/check'))
    ```

12. **os.path.isfile(path)**
 - Проверяет, является ли путь файлом.
    ```python
    print(os.path.isfile('/path/to/file'))
    ```

13. **os.path.isdir(path)**
 - Проверяет, является ли путь каталогом.
    ```python
    print(os.path.isdir('/path/to/directory'))
    ```

### Работа с переменными окружения

1. **os.environ**
- Словарь, содержащий переменные окружения.
   ```python
   print(os.environ['HOME'])
   ```

2. **os.getenv(key, default=None)**
- Возвращает значение переменной окружения, если она существует, иначе возвращает значение по умолчанию.
   ```python
   print(os.getenv('HOME'))
   ```

3. **os.putenv(key, value)**
- Устанавливает значение переменной окружения.
   ```python
   os.putenv('MY_VARIABLE', 'my_value')
   ```

4. **os.unsetenv(key)**
- Удаляет переменную окружения.
   ```python
   os.unsetenv('MY_VARIABLE')
   ```

### Работа с путями

1. **os.path.join(path, \*paths)**
- Объединяет один или несколько компонентов пути.
   ```python
   full_path = os.path.join('/path', 'to', 'file.txt')
   ```

2. **os.path.split(path)**
- Разделяет путь на кортеж (head, tail).
   ```python
   head, tail = os.path.split('/path/to/file.txt')
   ```

3. **os.path.splitext(path)**
- Разделяет путь на кортеж (root, ext).
   ```python
   root, ext = os.path.splitext('/path/to/file.txt')
   ```

4. **os.path.basename(path)**
- Возвращает имя файла или каталога.
   ```python
   filename = os.path.basename('/path/to/file.txt')
   ```

5. **os.path.dirname(path)**
- Возвращает путь к каталогу.
   ```python
   directory = os.path.dirname('/path/to/file.txt')
   ```
