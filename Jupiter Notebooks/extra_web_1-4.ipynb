##### индексация строк

```python
string = "Я хорошо отдыхаю на море, тут очень жарко."

# поделим строку на 3 части: 
# 1 - в обратном порядке через 1 
# 2 - в прямом подяке, но не вкл границы 
# 3 - в обратном порядке (используем отриц индексы) не вкл 2 символа с конца 

print(len(string))
len1 = (len(string) // 3)
len2 = (len(string) // 3) * 2
print(len1, string[len1])
print(len2, string[len2])

print(string[len1::-2])  # start end (не вкл) step 
print(string[len1 + 2:len2:1])
print(string[-2:len2:-1])

print(string[len1::-2] + string[len1 + 2:len2:1] + string[-2:len2:-1])
```

##### Именованные и позиционные аргументы

```python
def example_function(name, age, gender="F"):
	print(f"name={name}, age={age}, gender={gender}")


name = "Виктория"
age = 21
gender = "F"

# позиционные аргументы 
example_function(name, age, gender)
example_function("Виктория", 21, "F")  # порядок КРИТИЧЕСКИ важен 
example_function("Виктория", "F", 21)  # порядок КРИТИЧЕСКИ важен - тут неправильный 

# именованные аргументы
example_function(name="Виктория", gender="F", age=21)

# и те, и те аргументы
example_function("Виктория", gender="F", age=21)

# ошибки 
example_function(name='Александра', 26, 'F')  # именованные ДО позиционных 
example_function(26, 'F', name='Александра')  # 2 name 
```

##### * - разделяет позиционные и именованные аргументы 

```python
def example_function(*, name, age, gender="F"):
	print(f"name={name}, age={age}, gender={gender}")


name = "Виктория"
age = 21
gender = "F"

example_function(name=name, age=age, gender=gender)
example_function(name, age=age, gender=gender)
```

##### Распаковка аргументов

```python
def example_function(*args, **kwargs):  # запаковка
	print(type(args), args)
	print(type(kwargs), kwargs)
	for i in args:
		print(i)
	for key, value in kwargs.items():
		print(key, value)


# example_function(1, 2, 3, a=1, b=2, sasha=10)

lst = (1, 2, 3)
my_dict = {'a': 1}

example_function(*lst, **my_dict)  # 1, 2, 3, a=1 распаковка
```

##### Чистая функция 

1. **Неизменяемость (Immutability)**
Чистые функции не изменяют входные данные. Вместо изменения состояния объектов они возвращают новые объекты. Это свойство делает код более безопасным и легким для отладки, поскольку состояние системы остается неизменным между вызовами функций.

2. **Без побочных эффектов (No Side Effects)**
Чистые функции не влияют на внешний мир за пределами своих параметров и возвращаемого значения. Они не взаимодействуют с глобальным состоянием, не изменяют внешние переменные и не вызывают сторонних эффектов, таких как ввод-вывод, сетевые запросы или модификация базы данных.

3. **Предсказуемость (Determinism)**
Из-за отсутствия побочных эффектов и неизменяемости, результаты вызова чистой функции всегда одинаковы при одних и тех же входных данных. Это свойство делает тестирование и отладку кода гораздо проще, поскольку можно быть уверенным в том, что функция будет работать одинаково в разных средах и с разными данными.

4. **Пересчитываемость (Referential Transparency)**
Если функция является чистой, её результат можно заменить на её аргументы без изменения поведения программы. Это означает, что любое место в коде, где используется результат функции, может быть заменено непосредственным использованием этого результата, что упрощает оптимизацию и анализ кода.

```python
def add(a, b):
	return a + b


print(add(4, 5))
print(add(4, 5))
print(add(4, 5))
```

```python
def add(a=[]):
	a.append(5)
	return a


print(add())
print(add())
print(add())
```

```python
def add(a=None):
	a = []
	a.append(5)
	return a


print(add())
print(add())
print(add())
```

```python
a = []


def add():
	a.append(5)
	return a


print(add())
print(add())
print(add())
```

##### Рекурсия

```python
def factorial(n):
	# База рекурсии: факториал 0 равен 1
	if n == 0:
		return 1
	else:
		return n * factorial(n - 1)

	# 2 * factorial(1) -> 2 * (1 * factorial(0)) -> 2 * (1 * (1)) -> 2


# print(factorial(-1)) # тут переполнится стэк
print(factorial(0))
print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(10))
```

```python
def sum_until_zero(numbers):
	# Проверяем, является ли список пустым
	if not numbers:
		return 0

	# Берем первое число из списка
	first_number = numbers[0]

	# Если оно меньше или равно нулю, возвращаем 0
	if first_number <= 0:
		return 0

	# Иначе рекурсивно вызываем функцию для оставшейся части списка
	else:
		return first_number + sum_until_zero(numbers[1:])


numbers = [5, 3, 7, -1, 9, 6]
total = sum_until_zero(numbers)
print(total)
```

## 4 - Модули и пакеты
- Как импортировать файлы
- Пространство имен и области видимости

1. **Глобальная область видимости (Global Scope)**:
   - Переменные, объявленные на уровне модуля (файла), имеют глобальную область видимости.
   - Эти переменные доступны из любой части модуля.
   - Ключевое слово `global` используется для изменения глобальной переменной внутри функции.

.

2. **Локальная область видимости (Local Scope)**:
   - Переменные, объявленные внутри функции, имеют локальную область видимости.
   - Они доступны только в пределах этой функции.

.

3. **Область видимости вложенной функции (Enclosing Scope)**:
   - Это область видимости внешней функции для вложенной функции.
   - Переменные в этой области видимости доступны для всех вложенных функций.
   - Ключевое слово `nonlocal` используется для изменения переменной в области видимости внешней функции.

.

4. **Область видимости встроенных имен (Built-in Scope)**:
   - Эта область включает в себя все встроенные функции и объекты Python, такие как `len()`, `range()`, и так далее.

### Правило LEGB

Область видимости в Python управляется правилом LEGB, которое описывает порядок поиска переменных:
- **L (Local)**: Локальная область видимости текущей функции.
- **E (Enclosing)**: Область видимости внешних функций, если функция вложена.
- **G (Global)**: Глобальная область видимости текущего модуля.
- **B (Built-in)**: Встроенная область видимости.

##### Enclosing

```python
x = 5


def outer():
	x = "local"

	def inner():
		nonlocal x
		x = "nonlocal"
		print("inner:", x)

	inner()
	print("outer:", x)


outer()
print(x)
```

```python
x = 5


def outer():
	x = "local"

	def inner():
		global x
		x = "nonlocal"
		print("inner:", x)

	inner()
	print("outer:", x)


outer()
print(x)
```
